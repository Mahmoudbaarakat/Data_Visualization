## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
Query = """Select * from master""" 
df = pd.read_sql_query(Query,conn)

df_cleansed = df.dropna(subset=['ConvertedComp'])
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
plt.hist(df_cleansed['ConvertedComp'],bins=9,color='Skyblue',edgecolor='Black')
plt.title('Histogram of Converted Compensation')
plt.xlabel('Converted USD')
plt.ylabel('Frequency')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

df_cleansed_age = df.dropna(subset=['Age'])
plt.figure(figsize=(10,6))
sns.boxplot(x='Age',data = df_cleansed_age)
plt.title('Box plot of Age')
plt.xlabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
plt.figure(figsize=(10,6))
sns.scatterplot(x='Age',y='WorkWeekHrs',data=df_cleansed_age,color='blue',alpha=0.7)
plt.title('Scatter plot between Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

plt.figure(figsize=(10,6))
sns.scatterplot(x='WorkWeekHrs',y='CodeRevHrs',data=df_cleansed_age,hue='Age',size='Age',sizes=(10,100),palette='viridis',alpha=0.7)
plt.title('WorkWeekHrs VS CodeRevHrs by Bubble plot')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevhrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
Query = """Select * from DatabaseDesireNextYear"""
dff = pd.read_sql_query(Query,conn)

dff = dff.sort_values(by='Respondent', ascending=False)

# Select the top 5 databases
top_databases = dff.head(5)

# Create a pie chart using Seaborn aesthetics with Matplotlib
plt.figure(figsize=(8, 8))
sns.set_palette("pastel")
plt.pie(top_databases['Respondent'], labels=top_databases['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)

# Set title
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

filtered_age = df_cleansed_age[(df_cleansed_age['Age'] >= 30) & (df_cleansed_age['Age'] <= 35)]
plt.figure(figsize=(10, 6))

plt.bar(filtered_age['Age'], filtered_age['WorkWeekHrs'], color='blue', label='WorkWeekHrs')
plt.bar(filtered_age['Age'], filtered_age['CodeRevHrs'], color='orange', bottom=filtered_age['WorkWeekHrs'], label='CodeRevHrs')

# Set labels and title
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('Age')
plt.ylabel('Median Hours')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
